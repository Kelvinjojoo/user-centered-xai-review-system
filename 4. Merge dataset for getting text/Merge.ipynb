```python
import pandas as pd
```

```python
df_final = pd.read_csv("yelp_dataset_after_finalization.csv")
df_preproc = pd.read_csv("yelp_review_preprocessed.csv")

# Pastikan kolom review_id ada
assert "review_id" in df_final.columns, "Kolom review_id tidak ada di yelp_dataset_after_finalization.csv"
assert "review_id" in df_preproc.columns, "Kolom review_id tidak ada di yelp_review_preprocessed.csv"
```

```python
# ===============================
# 2. PILIH HANYA KOLUM YANG DIPERLUKAN DARI PREPROCESSED
# ===============================
# Pastikan kolom "text" ada
if "text" not in df_preproc.columns:
    raise ValueError("Kolom 'text' tidak ditemukan di yelp_review_preprocessed.csv")

df_preproc_small = df_preproc[["review_id", "text"]]
```

```python
# ===============================
# 3. MERGE BERDASARKAN review_id
# ===============================
df_merged = df_final.merge(
    df_preproc_small,
    on="review_id",
    how="left"    # pakai left biar tetap ikut semua baris dari final dataset
)
```

```python
# ===============================
# 4. CEK HASIL
# ===============================
print("Jumlah baris awal:", len(df_final))
print("Jumlah baris setelah merge:", len(df_merged))
df_merged.head()
```
